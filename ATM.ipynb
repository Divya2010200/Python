```python
class ATM:
    def __init__(self):
        self.balance=0
        self.__pin=0
        self.info()
    def info(self):
        input_value=int(input(""""1: To create the new pin
                                  2: To check the Balance
                                  3: To withdraw the amount
                                  4: To Deposit
                                  5: Mini Statement
                                  6: Reset Pin
                                  7: Exit : """))
        def reset_pin():
            check1=int(input("enter the pin: "))
            if check1==self.__pin:
                newPin1=int(input("Enter your new 4-digit pin: "))
                newPin2=int(input("Re-enter your new pin: "))
                if newPin1==newPin2:
                    print("Pin sucessfully saved")
                    self.__pin+=newPin2
                else:
                    print("Pin miss matched")
            else:
                  print("Pin miss matched")
            if input("b to go back e to exit: ")=="b":
                self.info()
            else:
                pass
            
        def creat_pin():
            if self.__pin==0:
                newPin1=int(input("Enter your new 4-digit pin: "))
                newPin2=int(input("Re-enter your new pin: "))
                if newPin1==newPin2:
                    print("Pin sucessfully saved")
                    self.__pin=newPin1
                else:
                    print("Pin miss matched")
            else:
                print("You already have a pin")
            if input("b to go back e to exit: ")=="b":
                self.info()
            else:
                pass
                
        def balance():
            check2=int(input("enter the pin: "))
            if check2==self.__pin:
                print("Your balance is {}".format(self.balance))
            else:
                print("Incoorect pin")
            if input("b to go back e to exit: ")=="b":
                self.info()
            else:
                pass
            
        def deposit():
            check=int(input("enter the pin: "))
            if check==self.__pin:
                howMuch=int(input("How much you want to diposit: "))
                how100=int(input("How many 100 notes: "))
                how200=int(input("How many 200 notes: "))
                how500=int(input("How many 500 notes: "))
                how2000=int(input("How many 2000 notes: "))
                if howMuch==100*how100+200*how200+500*how500+2000*how2000:
                    self.balance=self.balance+howMuch
                    print("Deposite Successfull")
                else:
                    print("Deposite Unsuccessfull")
            else:
                print("Enter the correct pin")
        def withraw():
            check3=int(input("enter the pin: "))
            if check3==self.__pin:
                howMuchWith=int(input("How much you want to withdraw: "))
                if howMuchWith<=self.balance:
                    self.balance=self.balance-howMuchWith
                    print("Withdraw Successfull")
                else:
                    print("Withdraw Unsuccessfull")
            else:
                print("Enter correct pin")
        
        
        if input_value==1:
            creat_pin()
        elif input_value==2:
            balance()
        elif input_value==3:
            withraw()
        elif input_value==4:
            deposit()
        elif input_value==6:
            reset_pin()
        else:
            pass 
```
